```python
import numpy as np
import xarray as xr
import os
import sys
from os.path import join,expanduser
import matplotlib.pyplot as plt
sys.path.append(os.path.abspath(os.path.join('..', '..', 'ECCO-v4-Python-Tutorial')))
import ecco_access as ea
user_home_dir = expanduser('~')
```

```python
path_all_vel = data_path + 'VEL/'
```

```python
file_names = []
if os.path.exists(path_all_vel) and os.path.isdir(path_all_vel):
    for entry in os.listdir(path_all_vel):
        full_path = os.path.join(path_all_vel, entry)
        if os.path.isfile(full_path):
            file_names.append(entry)
```

```python
ds_vels = xr.open_dataset(data_path + 'VEL/' + 'OCEAN_VELOCITY_mon_mean_1992-01_ECCO_V4r4_native_llc0090.nc')
ds_vels = ds_vels.sel(tile = 10).squeeze()

for f in range(1, len(file_names)):
    ds_temp = xr.open_dataset(path_all_vel + file_names[f])
    ds_temp = ds_temp.sel(tile = 10).squeeze()
    ds_vels = xr.concat([ds_vels, ds_temp], dim='time')
```

```python
ds_vels.to_netcdf(data_path + 'ecco_tile_11_vel')
```
